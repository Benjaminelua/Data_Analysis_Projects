```python
#Importing the libraries
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#reading CSV file
df = pd.read_csv(r"C:\Users\user\OneDrive\Desktop\gbemidebe.csv")
```

```python
df
```

```python
# Relationship between lewdness suggestion and religion
plt.figure(figsize=(8, 3.2))
ax = sns.countplot(x='LEWDNESS?', hue='RELIGION', data=df, palette='Paired')
plt.title('Lewdness Suggestion by Religion')
plt.xlabel('Suggests Lewdness')
plt.ylabel('Number of Respondents')
plt.legend(title='Religion')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 3), 
                textcoords='offset points')
plt.show()
```

```python
# Relationship between lewdness suggestion and gender
plt.figure(figsize=(8, 3.2))
ax = sns.countplot(x='LEWDNESS?', hue='GENDER', data=df, palette='Paired')
plt.title('Lewdness Suggestion by Gender')
plt.xlabel('Suggests Lewdness')
plt.ylabel('Number of Respondents')
plt.legend(title='Gender')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 3), 
                textcoords='offset points')
```
